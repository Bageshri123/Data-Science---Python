# Interrogating the dataframe
---

We use pandas to create dataframes from data sets in files or online tables.

Once we have the dataframe we might want to:  
*   display a subset of rows
*   *   we can do this using df.head(), df.tail() or df.iloc[index]
*   display a subset of columns
*   *   we can do this using the column name as the key, e.g.   
`                df['rain (mm)']`  
*   *   we can display more than one column, simply by listing the columns in the square brackets, e.g.   
`                df[['rain (mm)', 'tmin (degC)', 'sun (hours)']]`  
*   *   and can dislay a subset of rows for a subset of columns, e.g.  
 `               df[['rain (mm)', 'tmax (degC)']].head(10)`

### Exercise 8 - total net migration over 5 years
---

Using the Country Migration data sheet, display the record at position 10.  Using the column names to access the data in each column, add the five columns containing the net migration data together to calculate the total net migration over the 5 years between 2015 and 2019.

```python
import pandas as pd
url="https://github.com/Bageshri123/Data-Science---Python/blob/main/Data%20Sets/public_use-talent-migration%20(1)%20(3).xlsx?raw=true"
df = pd.read_excel(url,sheet_name="Country Migration")

#print the 10th record
df=df.iloc[9]
print(df)

#calculate net migration over the 5 years for 10th record
net_5yr_migration=df[['net_per_10K_2015','net_per_10K_2016','net_per_10K_2017','net_per_10K_2018','net_per_10K_2019']].sum()
print("\nNet migration between 2015 to 2019 per 10k for record 10 :",net_5yr_migration)
```

**Extra Challenge :
Using industry migration data sheet, display net migration durin 2015-2019 for each country and each industry**

```python
import pandas as pd

url="https://github.com/Bageshri123/Data-Science---Python/blob/main/Data%20Sets/public_use-talent-migration%20(1)%20(3).xlsx?raw=true"
df = pd.read_excel(url,sheet_name="Industry Migration")


for i in df['country_name'].unique():
  df1=df.loc[df['country_name']==i]

  for j in df1['industry_name'].unique():
    total_mig=0
    df2=df1.loc[df1['industry_name']==j]
    total_mig+=df2['net_per_10K_2015'].sum()+df2['net_per_10K_2016'].sum()+df2['net_per_10K_2017'].sum()+df2['net_per_10K_2018'].sum()+df2['net_per_10K_2019'].sum()
    print("\nThe total net migration into",i, "of skilled workers for the",j, "industry, during the years 2015-2019 was", total_mig)


  


```
