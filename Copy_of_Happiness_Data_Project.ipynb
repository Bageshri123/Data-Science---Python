# How happy is the world?

![HappinessImage-Benjamin Scott](https://drive.google.com/uc?id=1wWdXTclLAjPpZUiKkR8XGg0Sl7iD1u8T)  

(Image by Benjamin Scott, [source](https://www.natureindex.com/news-blog/data-visualization-these-are-the-happiest-countries-world-happiness-report-twenty-nineteen))   

The Sustainable Development Solutions Network (SDSN) collects data across the world relating to happiness.  They use this data to rank countries in order of happiness factor.

This is not an exact science but can give food for thought in terms of what factors might have the most impact on a nation's happiness levels.

Data is taken from the Gallup World Poll, so not collected directly by SDSN.  
Countries are grouped by region.  

### The factors included are:
Economy (measured in GDP per Capita)	
Family (support systems) 	
Health (measured by Life Expectancy)	
Freedom (sense of)  	 
Trust (Government Corruption)	
Generosity (charitable inclinations)  	
Dystopia Residual 
*  Dystopic is the theoretical most unhappy country with the lowest levels in all six of the above factors  
*  The Residual measure is a calculated as the average of the six distances from lowest

Let's take a look at the data


---
### Open a data set



```
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

url="https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2015.xlsx?raw=true"
happiness=pd.read_excel(url)

happiness.head()
```

---
### Sort the data in different ways

The data is currently sorted in order of rank.  To sort the data in the table, run the code below, which identifies the column on which to sort in the brackets.

Then, **try sorting on other columns** *Note: you must type the column heading in the quotes and exactly as it appears in the table (including capitalisation)*.



```
#sort in order of family
family_happiness=happiness.sort_values('Family')
family_happiness
```

```
#sort in order of health
health_happiness=happiness.sort_values('Health (Life Expectancy)')
health_happiness.head()
```

```
#sort in order of Generosity
generosity_happiness=happiness.sort_values('Generosity')
generosity_happiness.head()
```

---
### Visualising the data

There are many things we can do with this data and you will be asked to think about this.

For now, here is an example of how this data could be visualised in a scatter plot.  

You can plot a line graph by replacing *scatter* with *plot* AND if you want to plot more than one line on the same chart, then you can just add one line for each column to include (one column must be the same in all plots).

Try some different graphs, using different columns.

```
#scatter plot to display correlation between happiness rank and health

x=happiness['Happiness Rank']
y=happiness['Health (Life Expectancy)']

plt.scatter(x,y)
plt.xlabel('Happiness Rank')  
plt.ylabel('Health')

regression=linregress(x,y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope*x + intercept)

plt.show

```

```
#line graph to display correlation between happiness rank and health

x=happiness['Happiness Rank']
y=happiness['Health (Life Expectancy)']

plt.plot(x,y)
plt.xlabel('Happiness Rank')  
plt.ylabel('Health')

regression=linregress(x,y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope*x + intercept)


plt.show
```

```
#line graph to display correlation between happiness rank and health and happiness rank and family

x=happiness['Happiness Rank']
y=happiness['Health (Life Expectancy)']

plt.plot(x,y,label='Health')

regression=linregress(x,y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope*x + intercept)


x=happiness['Happiness Rank']
y=happiness['Family']

plt.plot(x,y,label='Family')

regression=linregress(x,y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope*x + intercept)

plt.xlabel('Happiness Rank')  

plt.legend()
plt.show()
```

```
#line graph to display correlation between happiness rank and Generosity and happiness rank and Freedom

x=happiness['Happiness Rank']
y=happiness['Generosity']


plt.plot(x,y,label='Generosity')


x=happiness['Happiness Rank']
y=happiness['Freedom']

plt.plot(x,y,label='Freedom')
plt.xlabel('Happiness Rank')  

plt.legend()
plt.show()
```

There are data sets for the years 2015 to 2019 available.  To access and try out other years, change 2015 to the required year in the URL in the first code cell.  Leave the rest exactly as it is.  

Other years may have different column headings and so there will be different data to play with.

```
#Read Happiness data for 2018
import pandas as pd
from scipy.stats import linregress
import matplotlib.pyplot as plt

url="https://github.com/futureCodersSE/working-with-data/blob/main/Happiness-Data/2018.xlsx?raw=true"
happiness18=pd.read_excel(url)

happiness18.head()
```

```
#boxplot of the social support score

data=happiness18['Social support']
plt.boxplot(data)
plt.xlabel('Social support') 
plt.show
```

```
#pie chart to show how much each factor counts towards Finland's happiness score
finland=happiness18.loc[happiness18['Country or region']=='Finland']

data=finland[['Freedom to make life choices','GDP per capita','Social support','Healthy life expectancy','Generosity','Perceptions of corruption']]
labels=data.keys()

plt.pie(data,labels=labels,autopct='%1.2f%%')
plt.title('Finland Happiness Factors')
plt.show()
```

```
#line graph to display correlation between happiness score and GDP per Capita
x=happiness18['Score']
y=happiness18['GDP per capita']

plt.plot(x,y)
plt.xlabel('Happiness Score')  
plt.ylabel('GDP per capita')

regression=linregress(x,y)
slope = regression.slope
intercept = regression.intercept
plt.plot(x, slope*x + intercept)


plt.show()
```

```
#scatter plot of happiness and perception of corruption score

x=happiness18['Score']
y=happiness18['Perceptions of corruption']

plt.scatter(x,y)
plt.xlabel('Happiness Score')  
plt.ylabel('Perceptions of corruption')

plt.show()



```
